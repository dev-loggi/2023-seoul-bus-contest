```python
# 1. 정류소ID, 정류소명 drop
loc.drop(['NODE_ID', '정류소명'], axis=1, inplace=True)

# 2. X좌표, Y좌표 각각 longitude, latitude로 이름 변경, ARS-ID는 버스정류장ARS번호로 이름 변경
loc.rename(columns={
    'ARS_ID': '버스정류장ARS번호',
    'X좌표': 'longitude',
    'Y좌표': 'latitude'
}, inplace=True)
```

```python
fn = pd.merge(loc, kk, on='버스정류장ARS번호', how='inner')
fn.shape
```

```python
#
fn['temp1'] ='('
fn['temp2'] = ')'
fn['버스정류장ARS번호'] = fn['버스정류장ARS번호'].astype('str')
fn['정류소명'] = fn['정류소'] + fn['temp1'] + fn['버스정류장ARS번호'] + fn['temp2']
```

```python
drop_cols = ['temp1', 'temp2', '정류소']
fn.drop(drop_cols, axis=1, inplace=True)
```
