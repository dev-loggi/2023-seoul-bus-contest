```python
import pandas as pd
```

```python
apt = pd.read_csv('/content/drive/MyDrive/시내버스_공모전_/jupyter notebook/0_금천구_시각화/data/서울시 공동주택 아파트 정보.csv', encoding='cp949')
```

```python
# 주소(시군구) = 금천구 만 추출
apt = apt[apt['주소(시군구)'] == '금천구']
```

```python
apt.head(2)
```

```python
apt.columns
```

```python
in_cols = [
    'k-아파트명',  '주소(읍면동)', '좌표X', '좌표Y'
]

lst = []
for i in apt.columns.to_list():
  if i not in in_cols:
    lst.append(i)

apt.drop(lst, axis=1, inplace=True)
```

```python
apt.head(2)
```

```python
# 이름 좀 바꾸자.
apt.rename(columns = {
    'k-아파트명' : '아파트',
    '주소(읍면동)': '주소',
    '좌표X': 'longitude',
    '좌표Y' : 'latitude'
}, inplace=True)
```

```python
apt.head(2)
```

```python
# okay 이제 시각화만 하면 된당!
# csv파일 저장이나 하자.
apt.to_csv('/content/drive/MyDrive/시내버스_공모전_/jupyter notebook/0_금천구_시각화/data/apt_시각화_최종.csv', index=False)
```
