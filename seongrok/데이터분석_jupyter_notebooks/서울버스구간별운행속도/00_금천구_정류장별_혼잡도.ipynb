# 4.속도 구하기

## 1) 간단한 계산

```python
route_target_nm = '600'
route_target_id = route_master.loc[route_master['노선_명칭'] == route_target_nm, '노선_ID'].values[0]

trip_time_1 = trip_time_all.loc[trip_time_all['노선_ID'] == route_target_id].copy()

# 날짜별 정류장구간별 운행시간_출근, 운행시간_퇴근 평균운행시간 계산
trip_time_1['운행시간_출근'] = trip_time_1.loc[:, ['운행시간_07시', '운행시간_08시']].mean(axis=1)
trip_time_1['운행시간_퇴근'] = trip_time_1.loc[:, ['운행시간_17시', '운행시간_18시']].mean(axis=1)
trip_time_1

# 전구간날짜 평균
trip_time_mean_1 = trip_time1.groupby(by=['노선_ID', '출발_정류장_ID', '도착_정류장_ID', '출발_정류장_순서', '도착_정류장_순서'], as_index=False)[['운행시간_출근', '운행시간_퇴근']].mean()
trip_time_mean_1.sort_values(by='도착_정류장_순서', ignore_index=True, inplace=True)
trip_time_mean_1.head()
```

```python
route_node_target = route_node_master.loc[route_node_master['노선_ID'] == route_target_id]

trip_speed_1 = pd.merge(route_node_target, trip_time_mean_1, left_on=['노선_ID', '정류장_ID', '정류장_순서'], right_on=['노선_ID', '도착_정류장_ID', '도착_정류장_순서'], how='left')
trip_speed_1.sort_values(by='정류장_순서', ignore_index=True, inplace=True)

trip_speed_1.iloc[0, 5] = trip_speed_1.iloc[1, 4]
trip_speed_1.iloc[0, 7] = trip_speed_1.iloc[1, 6]

trip_speed_1['운행속도_출근'] = trip_speed_1['링크_구간거리(m)'] / trip_speed_1['운행시간_출근'] * (3600 / 1000)
trip_speed_1['운행속도_퇴근'] = trip_speed_1['링크_구간거리(m)'] / trip_speed_1['운행시간_퇴근'] * (3600 / 1000)
trip_speed_1.iloc[0, 10] = 0.0
trip_speed_1.iloc[0, 11] = 0.0
trip_speed_1
```
